```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("we_data.csv")
df
```

```python
df.info()
```

```python
# Convert 'Date_Time' to datetime 
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Drop any rows with NaN values if necessary
df.dropna(inplace=True)
```

```python
# check the null values
df.isnull().sum()
```

```python
sns.scatterplot(x='temperature', y='humidity', data=df)
```

```python
# Create a line plot for temperature
plt.figure(figsize=(6, 6))
sns.lineplot(data=df, x=df.index, y='temperature', marker='o')
# Customize the plot
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid()
```

```python
#Explanation:

-This line plot visualizes the temperature changes over the recorded dates.
-The x-axis represents the date and time, while the y-axis shows temperature in degrees Celsius.
-Markers on the line help indicate specific data points.
```

```python
# Create a line plot for humidity
plt.figure(figsize=(7, 6))
sns.lineplot(data=df, x=df.index, y='humidity', marker='o', color='orange')
plt.title('Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Explanation:

-This line plot shows how humidity levels change over time.
-Similar to the temperature plot, it highlights the relationship between the date and humidity percentage.
```

```python
# Create a line plot for wind speed
plt.figure(figsize=(6, 6))
sns.lineplot(data=df, x=df.index, y='wind_speed', marker='o', color='green')
plt.title('Wind Speed Over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed (m/s)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.hist(df['wind_speed'].dropna(), bins=5, color='lightblue', edgecolor='black')
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (units)')
plt.ylabel('Frequency')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#Explanation:

-This histogram displays the distribution of recorded wind speeds.
-The x-axis represents wind speed values, while the y-axis shows how often each range of wind speed occurs.
-Its useful for understanding the general conditions regarding wind speed.
```

```python
# Calculate average temperature and humidity
average_temperature = df['temperature'].mean()
average_humidity = df['humidity'].mean()

print(f"Average Temperature: {average_temperature:.2f} C")
print(f"Average Humidity: {average_humidity:.2f} %")
```

```python
#Summary of Findings
-Temperature Trend: There are noticeable peaks and drops in temperature over time, suggesting changes in weather conditions.
-Humidity Levels: Humidity generally fluctuates, showing higher levels during certain periods.
-Wind Speed Distribution: The histogram illustrates how often different wind speeds are recorded, providing insights into typical weather patterns.
```
